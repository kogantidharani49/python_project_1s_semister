```python
import sqlite3      
import csv         
```

```python
con = sqlite3.connect("Housing.db")
cursor = con.cursor()


create_table = """CREATE TABLE housing_prices11(                               
                        price INTEGER,
                        area INTEGER,
                        bedrooms INTEGER,
                        bathrooms INTEGER,
                        stories INTEGER,
                        mainroad TEXT,
                        guestroom TEXT,
                        basement TEXT,
                        hotwaterheating TEXT,
                        airconditioning TEXT,
                        parking INTEGER,
                        prefarea TEXT,
                        furnishingstatus TEXT);
                """

cursor.execute(create_table)



with open("Housing_Price_Data.csv") as f:
    data = csv.reader(f)
    for row in data:
        if row[0]=="price":      
            continue
        cursor.execute("INSERT INTO housing_prices11 VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)", (row[0], row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[9], row[10], row[11], row[12]))



cursor.execute("SELECT * FROM housing_prices11").fetchall() 
```

```python
con.commit()
con.close()
```
